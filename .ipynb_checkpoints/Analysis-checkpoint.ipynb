```python
import os

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE

sns.set(rc={'axes.facecolor':'#00171d', 'figure.facecolor':'#00171d', 'grid.color': 'white', 'text.color': 'white', 
    'axes.labelcolor': 'white', 'grid.color': 'white', 'text.color': 'white', 'xtick.color': 'white', 'ytick.color': 'white'},
       font_scale=3)
```

```python
def load_files_to_matrix(normalise=True):    
    data = []
    for filename in os.listdir("data"):

        file_tags = filename.split("_")
        dict_values = dict_to_vector(np.load("data/" + filename).item())
        
        if normalise:
            dict_values = np.divide(dict_values, max(dict_values))
        data.append(dict_values)

    return np.asarray(data)

def dict_to_vector(dict_words):
    return np.array(list(dict_words.values()))
```

```python
# Needs a lot more data points.
tsne_embedding = TSNE(n_components=3).fit_transform(load_files_to_matrix())
fig = plt.figure(0, figsize=(20,20), dpi=150)
ax = fig.gca(projection='3d')
ax.scatter(tsne_embedding[:,0], tsne_embedding[:,1], tsne_embedding[:,2], s=10, c=sns.color_palette("inferno_r", tsne_embedding.shape[0]))
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.view_init(azim=200, elev=10)
ax.set_xticks([]) 
ax.set_yticks([]) 
ax.set_zticks([])
plt.show()
```

```python
tsne_embedding = TSNE(n_components=2).fit_transform(load_files_to_matrix())
fig = plt.figure(0, figsize=(20,20), dpi=50)
ax = plt.gca()
plt.scatter(tsne_embedding[:,0], tsne_embedding[:,1], c=['#%02x%02x%02x' % tuple([int(255*value) for value in color]) for color in sns.color_palette("inferno_r", tsne_embedding.shape[0])])
plt.plot()
```
